## 1.기쁨데이터 불러오기

```python
기쁨_부부 = []
for i in range(1,99):
  data_name = f'Empathy_기쁨_부부_{i:d}'
  file_path = f'/gdrive/MyDrive/NLP 공감형 대화 데이터셋/공감형 대화(부부, 연인)/VS_기쁨_부부/{data_name}.tsv'

  df = pd.read_csv(file_path, delimiter='\t')
  df['파일번호'] = i
  기쁨_부부.append(df)

기쁨_부부_combined = pd.concat(기쁨_부부, ignore_index=True)
```

```python
기쁨_연인 = []
for i in range(1,92):
  data_name = f'Empathy_기쁨_연인_{i:d}'
  file_path = f'/gdrive/MyDrive/NLP 공감형 대화 데이터셋/공감형 대화(부부, 연인)/VS_기쁨_연인/{data_name}.tsv'

  df = pd.read_csv(file_path, delimiter='\t')
  df['파일번호'] = 98+i
  기쁨_연인.append(df)

기쁨_연인_combined = pd.concat(기쁨_연인, ignore_index=True)
```

## 2. 기쁨 데이터셋 병합

```python
pleasure_df = pd.concat([기쁨_부부_combined, 기쁨_연인_combined], ignore_index=True)
pleasure_df['emotion']='pleasure'
```
